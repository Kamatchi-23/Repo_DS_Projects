<h4 class="list-group-item list-group-item-action active" >5.  MULTI VARIATE ANALYSIS, FEATURE ENGINEERING</h4>

>#### A. Write a program

    Step 1: Create a  condition = 'LoanStatus'== 'Completed'| 'LoanStatus' == 'Defaulted' |\'LoanStatus' == 'Chargedoff'

    Step2: Create a user define function  using condition and `LoanStatus` column.

    Hint: df['LoanStatus'] = df.apply(user define function , axis=1)
    
        Sample output : 
        
        
        Completed    168
        
        Defaulted     59

```python
def LoanStat(d):
    if d['LoanStatus']=='Completed':
        return 'Completed'
    if (d['LoanStatus']=='Defaulted')|(d['LoanStatus']=='Chargedoff'):
        return "Defaulted"
df['LoanStatus']=df.apply(LoanStat,axis=1)
df['LoanStatus'].value_counts()
```

#### B. Write a program

    Step 1: Create a  dictionary called categories = 1: 'Debt Consolidation', 2: 'Home Improvement', 3: 'Business', 6: 'Auto', 7: 'Other'

    Step2: Create a user define function  using categories and `ListingCategory (numeric)` column.

    Hint: df['ListingCategory (numeric)'] = df.apply(user define function , axis=1)
    
        Sample output : 
        
        
        Debt Consolidation    106

        Other                  65

        Business               25

        Home Improvement       22

        Auto                    9

```python
def ListCatCount(df):
    if not df['LoanStatus']==None:
        try:
            i={1: 'Debt Consolidation', 2: 'Home Improvement', 3: 'Business', 6: 'Auto',
                    7: 'Other'}[df['ListingCategory (numeric)']]
        except:
            i='Other'
    else: i=None
    return i

df['ListingCategory (numeric)']=df.apply(ListCatCount,axis=1)
df['ListingCategory (numeric)'].value_counts()
```

#### C. Display the box plot for `ProsperRating (Alpha)` vs `LoanOriginalAmount` and hue = `Loan Status`[Completed,Defaulted]. Write your observations

```python
plt.figure(figsize=(10,7))
sns.set(font_scale=1)
sns.boxplot(x=df['ProsperRating (Alpha)'],y=df['LoanOriginalAmount'],hue=df['LoanStatus'],showmeans=True)
plt.legend(loc='upper right')
plt.show()

#Observations:
    #Loan amount has been the highest for Completed Loans and Prosper rating A & C
    #For Prosper Rating A, B, there are hardly any defaulted cases
    #Prosper Rating(except for HR) with respect to loan amount are majorly positively skewed
```
